# Triplet Visualization

```python
# import csv with triplets
header = 1

root = []
metaA = []
metaB = []

import csv
reader = csv.reader(open('20200830_MetaTriplet(Zwischenexport).csv', 'r', encoding ="utf-8"),delimiter=';')
for row in reader:
    if header:
        header = 0
        continue
    r,A,B = row
    root.append(r)
    metaA.append(A)
    metaB.append(B)
```

```python
import networkx as nx 
import matplotlib.pyplot as plt 
%matplotlib inline

G = nx.Graph()
nodeA = metaA
nodeB = metaB
relation = root

for a, b in zip(nodeA, nodeB):
    G.add_edge(a, b)

pos = nx.spring_layout(G, k=0.15, iterations=20)
plt.figure(figsize=(12,12))
nx.draw(G,pos,width=1,linewidths=1,node_size=500,alpha=0.9,labels={node:node for node in G.nodes()})
labels = dict(zip(list(zip(nodeA, nodeB)),relation))
nx.draw_networkx_edge_labels(G,pos,edge_labels=labels,font_color='red')

plt.axis('off')
plt.show()
```
